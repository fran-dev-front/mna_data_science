Melt convierte un Dato en un formato donde una o mas columnas son variables de indentificacion por lo tanto cambia las columnas a filas, un pivot cambia los datos de las filas en columnas y las columnas en filas es similar a un group by si defines los parametros. El stack agrega dimensiones a las filas. No es lo mismo convertir una Columnas en fila que agrupar culumnas en una sola fila. Todas cambian las columnas a filas o al reves. Eligiria un stack() de una sola dimension.
